```python

# Identify numerical columns
numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

# Handle outliers using IQR method
for col in numerical_columns:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python

SS_X  = SS.fit_transform(df.iloc[:,1:16])
SS_X
# spliting the data in x and y
X = SS_X
Y = df.iloc[:,16]
```

# Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

# Implementing K-Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsClassifier

# Create a KNeighborsClassifier object
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
# Fit the classifier to the training data
knn.fit(X_train, Y_train)

# Predict the labels for the training data
Y_train_pred = knn.predict(X_train)

# Predict the labels for the test data
Y_test_pred = knn.predict(X_test)
```

```python
# Calculate the accuracy for the training data
train_accuracy = knn.score(X_train, Y_train)

# Calculate the accuracy for the test data
test_accuracy = knn.score(X_test, Y_test)

# Print the training and test accuracies
print("Training Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
```

```python

from sklearn.metrics import log_loss
Training_loss = log_loss(Y_train,Y_train_pred)
Test_loss = log_loss(Y_test,Y_test_pred)

# Print the training and test accuracies
print("Training_loss:", Training_loss)
print("Test_loss:", Test_loss)
```

1. What are the key hyperparameters in KNN?

In K-Nearest Neighbors (KNN), several hyperparameters significantly affect the model's performance. Here are the key ones:

Number of Neighbors (k): Determines how many neighbors to consider when making a prediction. Too low a k can lead to overfitting,
while too high a k can lead to underfitting.

Distance Metric: Defines how the distance between points is calculated. Common metrics include Euclidean distance, Manhattan distance, 
and Minkowski distance.

Weights: Dictates whether all neighbors contribute equally to the prediction. Options are:

Uniform: All neighbors have equal weight.

Distance: Closer neighbors have more influence.

Algorithm: Determines the algorithm used to compute nearest neighbors. Options include:

BallTree: Suitable for large datasets.

KDTree: Efficient for low-dimensional data.

Brute: Directly computes all pairwise distances (best for small datasets).

Leaf Size: Affects the speed and memory requirements of BallTree and KDTree algorithms. Smaller sizes may lead to faster queries but require more memory.

p (Power Parameter): Used with Minkowski distance. When p=1, it corresponds to Manhattan distance, and p=2 corresponds to Euclidean distance.

2. What distance metrics can be used in KNN?


K-Nearest Neighbors (KNN) can use several distance metrics to determine the “closeness” between data points. Here are some common ones:

Euclidean Distance: The straight-line distance between two points in Euclidean space. It’s the most commonly used metric.

Manhattan Distance (L1 norm): The sum of the absolute differences of their Cartesian coordinates. Like navigating a grid of city blocks.

Minkowski Distance: A generalization of Euclidean and Manhattan distances. It includes a parameter p, where p=1 is Manhattan and p=2 is Euclidean.

Chebyshev Distance: The maximum distance in any dimension. Good for chessboard-like problems where moves are along axes.

Hamming Distance: Used for categorical variables, it counts the number of positions at which the corresponding elements differ.

Cosine Similarity: Measures the cosine of the angle between two vectors, useful for text classification and other high-dimensional spaces.

Mahalanobis Distance: Takes into account the correlations of the data set and is scale-invariant. Great for multivariate data.
